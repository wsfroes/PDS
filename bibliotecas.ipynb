# Dissecando o código
## Bibliotecas

1. NumPy (import numpy as np)

Esta é a espinha dorsal de qualquer cálculo numérico em Python. O Python puro é lento para matemática pesada porque ele trata cada número individualmente. 

O NumPy (Numerical Python) trabalha com vetores e matrizes de forma vetorizada.

Objetivo aqui: 

* Quando criamos t = np.arange(0, duracao, 1/Fs), não estamos criando uma lista simples; estamos gerando um vetor de alta performance que armazena milhares de pontos de tempo simultaneamente. 

* Toda a matemática de ondas senoidais (np.sin) é feita sobre esses vetores, o que é ordens de grandeza mais rápido do que usar loops for tradicionais.

2. Matplotlib (import matplotlib.pyplot as plt)

O computador processa números, mas o cérebro humano processa padrões visuais. 

O Matplotlib é o padrão da indústria para transformar dados numéricos em representações gráficas.

Objetivo aqui: 

* Ele atua como nossa "janela de inspeção". Sem ele, teríamos apenas uma lista de números flutuantes na tela. 

* Ele é responsável por renderizar os eixos, as cores, as legendas e a grade, transformando a matriz de dados em um gráfico interpretável (o osciloscópio digital).

3. SciPy FFT (from scipy.fft import fft, fftfreq)

O SciPy é a biblioteca irmã do NumPy, voltada para algoritmos de engenharia e ciência. 

O fft (Fast Fourier Transform) é o algoritmo que permite converter o sinal do domínio do tempo para o da frequência de forma eficiente.

Objetivo aqui:
 * fft: Executa a Transformada Rápida de Fourier. É o motor que identifica quais frequências existem dentro do seu sinal capturado.
 
 * fftfreq: Esta função é frequentemente esquecida, mas é vital. O fft te devolve apenas os valores de magnitude. O fftfreq calcula matematicamente o eixo X (as frequências correspondentes em Hertz) com base na taxa de amostragem ($Fs$) e no número de amostras ($N$). Sem ele, você teria um gráfico, mas não saberia se o pico estava em 60 Hz ou 600 Hz.

4. SciPy Signal Windows (from scipy.signal.windows import hamming)

Aqui entramos na engenharia de precisão. Quando capturamos um sinal em um tempo finito (nossos 100 milissegundos), o computador "corta" o sinal abruptamente. 

Isso gera um erro matemático chamado vazamento espectral (spectral leakage).

Objetivo aqui: 

* A função hamming gera uma curva de atenuação (uma janela) que suaviza o início e o fim da amostra, fazendo com que o sinal "desça a zero" nas bordas. Isso impede que o computador interprete o corte abrupto como uma frequência falsa. É um recurso essencial para qualquer análise espectral que exija credibilidade.

### O Fluxo de Dados
O processo completo funciona como uma linha de montagem:

* NumPy gera o mundo físico (o tempo e a onda).
* SciPy.signal condiciona os dados (aplica a janela de Hamming para limpar o ruído do corte).
* SciPy.fft extrai a informação do DNA do sinal.
* Matplotlib projeta o diagnóstico final para o seu monitor.
